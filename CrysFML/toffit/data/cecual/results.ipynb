```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
```

```python
file_name = 'cecual.xys'
x, y_obs, sy_obs, y_calc = np.loadtxt(file_name, skiprows=6, unpack=True)
```

```python
#plt.xlim(16000.0, 17000.0)
plt.plot(x, y_obs, 'o', x, y_calc, '-')
plt.xlabel("TOF")
#plt.ylabel("Intensity")
plt.title("toffit output for the data from POLARIS at ISIS on CeCuAl3")
plt.show()
```
