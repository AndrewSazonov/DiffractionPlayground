```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
file_name = 'powder_pattern.dat'

with open(file_name) as f:
    file_content = f.read().splitlines()

x_start, x_step, x_stop = np.fromstring(''.join(file_content[1]), sep=' ')
x = np.arange(x_start, x_stop + x_step, x_step)

y = np.fromstring(' '.join(file_content[2:]), sep=' ')
```

```python
plt.plot(x,y)
plt.show()
```
