```python
file_name = 'dhkl.dat'
x, y_sim, sy_sim, n = np.loadtxt(file_name, comments='#', unpack=True)
```

```python
x_mult = 3510
y_mult = 1e+2
y_add = 1

x = x * x_mult
y_sim = y_sim * y_mult + y_add
sy_sim = sy_sim * y_mult + y_add

header = """XYDATA McStas simulated data for BEER at ESS on duplex sample
  ...
  ...
  ...
INTER   1.0   1.0   0  0
TEMP  300.0"""
np.savetxt('duplex.dat', np.column_stack([x, y_sim, sy_sim]), fmt='%11.3f%9i%9i', header=header, comments='')
```

```python
plt.figure(1)
plt.plot(x, y_sim)
plt.xlabel("d-spacing (Ã…)")
plt.title("McStas simulated pattern for BEER (multiplied)")
plt.show()
```
